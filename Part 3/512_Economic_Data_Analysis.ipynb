```python
# plot the data points of smoke estime
import matplotlib.pyplot as plt

plt.figure(figsize = (20,12))
plt.plot(df['Year'], df['SmokeEstimates'],marker='o', color='purple', linewidth=2, linestyle='-', label='Avg Smoke Estimates')

plt.title('Avg smoke estimates per year', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Year', fontsize=14, fontweight='bold', color='violet')
plt.ylabel('Smoke Estimates', fontsize=14, fontweight='bold', color='violet')

plt.grid(True,color='Lightgrey')
```

```python
# Since the economic data we have is from 2013, hence we filter the dataset for years after 2013 and then we plot it

df_2018_onwards = df[df['Year'] >= 2013]

plt.figure(figsize = (20,12))
plt.plot(df_2018_onwards['Year'], df_2018_onwards['SmokeEstimates'],marker='o', color='purple', linewidth=2, linestyle='-', label='Avg Smoke Estimates')

plt.title('Avg smoke estimates per year', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Year', fontsize=14, fontweight='bold', color='violet')
plt.ylabel('Smoke Estimates', fontsize=14, fontweight='bold', color='violet')

plt.grid(True,color='Lightgrey')
```

```python
# Occupations in Fire Fighting & Prevention vs. Avg Smoke Estimates graph plotting steps

df_occupations = pd.read_csv('Occupations.csv')

target_occupation1 = "Fire Fighting & Prevention, & Other Protective Service Workers Including Supervisors"

df_occupations1 = df_occupations[df_occupations['Occupation'] == target_occupation1]

target_occupation2 = "Farming, Fishing, & Forestry Occupations"

df_occupations2 = df_occupations[df_occupations['Occupation'] == target_occupation2]

plt.figure(figsize = (10,6))
plt.plot(df_occupations1['Year'], df_occupations1['Share'], marker='o', color='pink', linewidth=2, linestyle='-', label='Fire Fighting & Prevention Occupations')
plt.plot(df_occupations2['Year'], df_occupations2['Share'], marker='o', color='violet', linewidth=2, linestyle='-', label='Farming, Fishing, & Forestry Occupations')
plt.plot(df_2018_onwards['Year'], df_2018_onwards['SmokeEstimates'],marker='o', color='purple', linewidth=2, linestyle='-', label='Avg Smoke Estimates')

plt.title('Occupations per year', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Year', fontsize=14, fontweight='bold', color='black')
plt.ylabel('Scale', fontsize=14, fontweight='bold', color='black')
plt.legend()

plt.grid(True,color='Lightgrey')


merged_df1 = pd.merge(df_2018_onwards, df_occupations1, how='inner', left_on='Year', right_on='Year')
correlation_coefficient1 = merged_df1['SmokeEstimates'].corr(merged_df1['Share'])
print(f"Correlation Coefficient: {correlation_coefficient1}")


merged_df2 = pd.merge(df_2018_onwards, df_occupations2, how='inner', left_on='Year', right_on='Year')
correlation_coefficient2 = merged_df2['SmokeEstimates'].corr(merged_df2['Share'])
print(f"Correlation Coefficient: {correlation_coefficient2}")

```
